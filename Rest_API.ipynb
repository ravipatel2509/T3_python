# REST_API(jsonplaceholder)

```python
import requests,json
import pandas as pd
url="https://jsonplaceholder.typicode.com/posts/1"
response=requests.get(url)
print(response)#object
print(response.status_code)
```

```python
print(response.text)
print(type(response.text))
```

```python
data=response.json()
print(data)#json datatype dict hase 
print(data["userId"])
print(data["id"])
print(data["title"])
```

pretty json format

```python
formatted_json=json.dumps(data,indent=4)
print(formatted_json)
```

# get multiple post

```python
url="https://jsonplaceholder.typicode.com/posts"
response=requests.get(url)
posts=response.json()
print(type(posts))#type of the post
print(len(posts))#length of the post
```

```python
#first 5 post id and title
for post in posts[:5]:
    print(post["id"])
    print(post["title"])
    print("-"*50)
    
```

# send query paramter


```python
#userid 1 hoy tevi post joie 
param={"userId":1}
url="https://jsonplaceholder.typicode.com/posts"
response=requests.get(url,params=param)#url,keyword=variable
filtered_post=response.json()
print(len(filtered_post))
```

```python
print(json.dumps(filtered_post,indent=5))
```

# send custom header

```python
headers={"user-Agent":"Mozilla/5.0","Accept":"application/json"}#request specific browser mention karyu hoy tena par thi ja karvu pade (format,browser)
response=requests.get(url,headers=headers)
print(response.status_code)#OK status code return kare che 
```

# 4 methods (offline code (MCQ))


**Send post Request**

```python
# Send post Request add the data by user 
```

```python
url="https://jsonplaceholder.typicode.com/posts"
new_post={"title":"Learning requets","body":"this is tutorial content","userId":1}
response=requests.post(url,json=new_post)
print(response.status_code)#new data post karie che ane te succesfully thaie gayu hoy to teno status code 200 che 
print(response.json())
```

**send put request** 

```python
# post ne update karva mate put use thaie
```

```python
url="https://jsonplaceholder.typicode.com/posts/1"
updated_post={"id":1,"title":"Updated title","body":"Updated body","userID":1}
response=requests.put(url,json=updated_post)
print(response.status_code)
print(response.json())
```

**send Patch request**

```python
url="https://jsonplaceholder.typicode.com/posts/1"
patch_data={"title":"Patched title"}
response=requests.patch(url,json=patch_data)#1 jaa parmeter no data change kari sakie 
print(response.status_code)
print(response.json())
```

**send Delete request**

```python
url="https://jsonplaceholder.typicode.com/posts/1"
response=requests.delete(url)
print(response.status_code)
```

**downlode image using binary Content**

```python
image_url="https://picsum.photos/200"
response=requests.get(image_url)
with open("random_image.jpg","wb")as f:#random image genrate kare
    f.write(response.content)
```

**saved json data to file**

```python
url="https://jsonplaceholder.typicode.com/posts/1"#json ni file bani ne ema data store kare 
response=requests.get(url)
data=response.json()
with open("post1.json","w")as f:
    json.dump(data,f,indent=4)
    
```

**read json data from file**

```python
with open("post1.json","r")as f:
    loaded_data=json.load(f)
print(loaded_data)
```

**convert API response to DataFrame**

```python
#json data ne dataframe maa convert kare
url="https://jsonplaceholder.typicode.com/posts"
response=requests.get(url)
posts=response.json()
df=pd.DataFrame(posts)
df.head()
```

**convert json string to python object**

```python
json_string='{"name":"alice","age":25}'
python_data=json.loads(json_string)
print(python_data)
print(type(python_data))
```

**convert python object into json string**

```python
person={"name":"alice","age":25}
json_output=json.dumps(person,indent=4)
print(json_output)
```

# pagination

```python
import requests
url="https://jsonplaceholder.typicode.com/posts"
param={"_page":1,"_limit":5}
response=requests.get(url,params=param)
data=response.json()
print("Post on page 1")
for i in data:
    print(i["id"],":",i["title"])
    
```
