```python
website="https://subslikescript.com/movies"
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify)
```

```python
Box=soup.find("article",class_="main-article")
movie_list=Box.find_all("a")
movie_title=[]
for i in movie_list:
    movie_title.append(i.text.strip())
movie_title
```

```python
#movie naa name ni file banvie joie ane eni te movie ni trascript e file maa save thavie joie 
links=[]
for i in Box.find_all("a",href=True):
    links.append(i["href"])
links
```

```python
root="https://subslikescript.com"
for i in links:
    website=root+i
    result=requests.get(website)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    transcript=soup.find("div",class_="full-script").text
    T=soup.find("h1").text.strip()
    with open(T+".txt","w",encoding="utf-8")as file:
        file.write(transcript)
    print("completed")
    
```

```python
import requests
from bs4 import BeautifulSoup

website = "https://www.bbc.com/news"
result = requests.get(website)
soup = BeautifulSoup(result.text, 'html.parser')
headings = soup.find_all("h2") 

headline_list = []
for i in headings:
    headline_list.append(i.text.strip())

print(headline_list) 
```

```python
#leptoimport requests
import requests
from bs4 import BeautifulSoup

website = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
result = requests.get(website)
soup = BeautifulSoup(result.text, 'html.parser')
name=soup.find_all("div",class_="card thumbnail")
lname= []
Pname= []
for i in name:
    lname.append(i.find('a').text.strip())
    Pname.append(i.find('h4').text.strip())
data={"lname":lname,"Pname":Pname}
import pandas as pd
df=pd.DataFrame(data)
df.to_csv("leptop.csv",index=False)
```
