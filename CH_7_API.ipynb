```python
import requests,json
API_Key='b1d5b44e2db41475cbd03b78525d2412'
Lat=str(23.0215374)
Lon=str(72.5800568)
website='https://api.openweathermap.org/data/2.5/weather?lat='+Lat+'&lon='+Lon+'&appid='+API_Key
result=requests.get(website)
coor=result.json()
print(json.dumps(coor,indent=5))
# currenttamp,pressur,humenity
currentTemp=coor['main']['temp']
pressure=coor['main']['pressure']
hum=coor['main']['humidity']
visibility=coor['visibility']
description=coor['weather'][1]['description']
speed=coor['wind']['speed']
country=coor['sys']['country']

print('Temp:',currentTemp)
print('Pressure :',pressure)
print('humidity :',hum)
print('visibility: ',visibility)
print('description: ',description)
print('speed :',speed)
print('Country :',country)
```

# Air Polution Data

```python
import requests,json
API_Key='b1d5b44e2db41475cbd03b78525d2412'
# air Qultiy index
Lat=str(23.0215374)
Lon=str(72.5800568)
website='http://api.openweathermap.org/data/2.5/air_pollution?lat='+Lat+'&lon='+Lon+'&appid='+API_Key
result=requests.get(website)
coor1=result.json()
print(json.dumps(coor1,indent=5))
aqi=coor1['list'][0]['main']['aqi']
print('aqi',aqi)
quality_indices={1:'Good',2:'fair',3:'moderate',4:'poor',5:'Very poor'}
print('current Air Quality :',quality_indices[aqi])
```

```python
#5day/3 hour 40semple
# date&Time,temp,pressure,humidity,disreption
import requests,json
API_Key='b1d5b44e2db41475cbd03b78525d2412'
# air Qultiy index
Lat=str(23.0215374)
Lon=str(72.5800568)
website="https://api.openweathermap.org/data/2.5/forecast?lat="+Lat+"&lon="+Lon+"&appid="+API_Key

result=requests.get(website)
coor2=result.json()
# print(json.dumps(coor2,indent=5))
# date=[]
main_data=coor2['list']
# print(json.dumps(coor2,indent=5))
d=[]
t=[]
p=[]
h=[]
D=[]
for i in main_data:
    d.append(i['dt_txt'])
    t.append(i['main']['temp'])
    p.append(i['main']['pressure'])
    h.append(i['main']['humidity'])
    D.append(i['weather'][0]['description'])
import pandas as pd

data={"Date &Time":d,"Temprature":t,"Pressure":p,"Humidity":h,"Discreption":D}
df=pd.DataFrame(data)
df
```
