```python
#Mercedes name /maileage/price/rating
with open("Certified used Mercedes-Benz for sale.html","r",encoding="utf-8")as file:
    content=file.read()
soup=BeautifulSoup(content,"html.parser")
Mercedes=soup.find_all("div",class_="vehicle-details")
Mname=[]
mailage=[]
price=[]
rating=[]
for i in Mercedes:
    Mname.append(i.find("h2").text.strip())
    mailage.append(i.find("div",class_="mileage").text.strip())
    price.append(i.find("span",class_="primary-price").text.strip())
    R=i.find("span",class_="sds-rating__count")
    if R:
        rating.append(R.text.strip())
    else:
        rating.append("data is not available")
data={"Mname":Mname,"Mailage":mailage,"price":price,"rateing":rating}
import pandas as pd
df=pd.DataFrame(data)
df.to_csv("Mercedes.csv",index=False)
df
```
