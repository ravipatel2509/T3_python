```python
website="https://subslikescript.com/movies"
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
```

```python
box=soup.find("article",class_='main-article')
movie_list=box.find_all("a")
movie_title=[]
for i in movie_list:
    movie_title.append(i.text.strip())
movie_title
```

```python
links=[]
for i in box.find_all('a',href=True):
    links.append(i['href'])
links
```

```python
root="https://subslikescript.com"
for i in links:
    website=root+i
    result=requests.get(website)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    transcript=soup.find('div',class_='full-script').text
    T=soup.find("h1").text.strip()
    with open(T+".txt",'w',encoding='utf-8') as file:
        file.write(transcript)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
website="https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
```

```python
laptops=soup.find_all("div",class_='card thumbnail')
laptop_names=[]
laptop_price=[]
for i in laptops:
    laptop_names.append(i.find('a').text.strip())
    laptop_price.append(i.find('h4').text.strip())
data={"laptop_names":laptop_names,'laptop_price':laptop_price}
import pandas as pd
df=pd.DataFrame(data)
df.to_csv("laptop.csv",index=False)
```

**30 most popular movie**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("30 Most Popular Movies.html","r",encoding='utf-8') as file:
    content=file.read()
soup=BeautifulSoup(content,'html.parser')
```

```python
Movie_name=[]
Rating=[]
Movies=soup.find_all('div',class_='article_movie_title')
for i in Movies:
    Movie_name.append(i.find('a').text.strip())
    Rating.append(i.find('span',class_='tMeterScore').text.strip())
data={"Movie_name":Movie_name,'Rating':Rating}
df=pd.DataFrame(data)
df
```

**Certified used Mercedes-Benz for sale**

```python
with open("Certified used Mercedes-Benz for sale.html","r",encoding='utf-8') as file:
    content=file.read()
soup=BeautifulSoup(content,'html.parser')
```

```python
car_data=soup.find_all('div',class_='vehicle-details')
Name=[]
Mileage=[]
Price=[]
Rating=[]
for i in car_data:
    Name.append(i.find('h2').text.strip())
    Mileage.append(i.find('div',class_='mileage').text.strip())
    Price.append(i.find('span',class_='primary-price').text.strip())
    R=i.find('span',class_='sds-rating__count')
    if R:
        Rating.append(R.text.strip())
    else:
        Rating.append("data not available")
data={"Name":Name,'Mileage':Mileage,'Price':Price,'Rating':Rating} 
df=pd.DataFrame(data)
df
```

**Flipkart**

```python
with open("flipkart.html","r",encoding="utf-8") as file:
    content=file.read()
soup=BeautifulSoup(content,'html.parser')  
box=soup.find_all('div',class_="_3pLy-c row")
mobile_name=[]
rating=[]
discount_price=[]
actual_price=[]
for i in box:
    mobile_name.append(i.find('div',class_='_4rR01T').text.strip())
    rating.append(i.find('div',class_='_3LWZlK').text.strip())
    discount_price.append(i.find('div',class_='_30jeq3 _1_WHN1').text.strip())
    a=i.find('div',class_='_3I9_wc _27UcVY')
    if a:
        actual_price.append(a.text.strip())
    else:
         actual_price.append("Missing")
data={"mobile_name":mobile_name,"rating":rating,"discount_price":discount_price,"actual_price":actual_price}
import pandas as pd
df=pd.DataFrame(data)
df
```
